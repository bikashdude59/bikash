```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
```

```python
def maxr2_score(regr,x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.15,random_state=r_state)
        regr.fit(x_train,y_train)
        y_pred=regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print('r2 score corresponding to:',r_state,"is:",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print ("max r2 score corresponding to",final_r_state,'is',max_r_score)
    return final_r_state

```

```python
lrge=linear_model.LinearRegression()
r_state=maxr2_score(lrge,x,y)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
neighbors={'n_neighbors':range(1,30)}
knr=KNeighborsRegressor()
gknr=GridSearchCV(knr,neighbors,cv=10)
gknr.fit(x,y)
gknr.best_params_
```

```python
knr=KNeighborsRegressor(n_neighbors=29)
r_state=maxr2_score(knr,x,y)
```

```python
from sklearn.model_selection import cross_val_score
print("mean r2 score for Linear Regression :",cross_val_score(lrge,x,y,cv=5,scoring='r2').mean())
print("Standard Deviation in r2 score for Linear Regression:",cross_val_score(lrge,x,y,cv=5,scoring='r2').std())
print()
print("mean r2 score for KNN Regression:",cross_val_score(knr,x,y,cv=5,scoring='r2').mean())
print("Standard Deviation in r2 score for KNN Regression:",cross_val_score(knr,x,y,cv=5,scoring='r2').std())
```

```python
from sklearn.linear_model import Lasso
lsreg=Lasso()
parameters={"alpha":[0.001,0.01,0.1,1]}
clf=GridSearchCV(lsreg,parameters,cv=10)
clf.fit(x,y)
clf.best_params_
```

```python
lsreg=Lasso(0.01)
```

```python
r_state=maxr2_score(lsreg,x,y)
```

```python
#using cross validation score with Lasso

print("Mean r2 score for Lasso Regression:", cross_val_score(lsreg,x,y,cv=5,scoring="r2").mean())
print("Statndard Deviation inn r2 score for Lasso Regression:",cross_val_score(lsreg,x,y,cv=5,scoring="r2").std())

#based on below output i prefered Linear regression.
```

```python
# using bossting tecgnique for boosting the score.
import warnings
warnings.filterwarnings("ignore")
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor()
parameters={"learning_rate":[0.001,0.01,0.1,1],"n_estimators":[10,100,500,1000]}
clf=GridSearchCV(gbr,parameters,cv=5)
clf.fit(x,y)
clf.best_params_

```

```python
gbr=GradientBoostingRegressor(learning_rate=0.01,n_estimators=100)
print("Mean r2 score for gradient boosting regressor:", cross_val_score(gbr,x,y,cv=5,scoring="r2").mean())
print("Statndard Deviation inn r2 score for gradient boosting regressor:",cross_val_score(gbr,x,y,cv=5,scoring="r2").std())
```

```python
#now use adaboost technique
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
ada_reg=AdaBoostRegressor()
parameters={"learning_rate":[0.001,0.01,0.1,1],"n_estimators":[10,100,500,1000],"base_estimator":[lrge,DecisionTreeRegressor()]}
clf=GridSearchCV(ada_reg,parameters,cv=5)
clf.fit(x,y)
clf.best_params_
```

```python
ada_reg=AdaBoostRegressor(base_estimator=lrge,learning_rate=0.1,n_estimators=10)
print("Mean r2 score for ada boost regressor:", cross_val_score(ada_reg,x,y,cv=5,scoring="r2").mean())
print("Statndard Deviation inn r2 score for ada boost regressor:",cross_val_score(ada_reg,x,y,cv=5,scoring="r2").std())
```

```python
r_state=maxr2_score(ada_reg,x,y)
```

```python
#till now ada boost regressor has best value.
#random state corresponding to 50 has highest r2.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.15,random_state=50)
ada_reg=AdaBoostRegressor()
ada_reg.fit(x_train,y_train)
y_pred=ada_reg.predict(x_test)
```

```python
print("r2score is :", r2_score(y_test,y_pred))
print('RMSE is :',np.sqrt(mean_squared_error(y_test,y_pred)))
```
