```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.21,random_state=45)
```

```python
max_r_score=0
```

```python
for r_state in range(42,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.21,random_state=r_state)
    regr=linear_model.LinearRegression()
    regr.fit(x_train,y_train)
    y_pred=regr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print('r2 score corresponding to random state:', r_state,"is:",r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(linear_model.LinearRegression(),x,y,cv=5,scoring='r2').mean()
```

```python
#we got max r2 score at random state 72
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=72)
lr=linear_model.LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
print("r2score is :", r2_score(y_test,y_pred))
print('RMSE is :',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
#saving the model
from sklearn.externals import joblib
joblib.dump(lr,'lr_baseball.obj')
```
